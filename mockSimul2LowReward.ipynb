```python
import mockSQLenv_low_reward as mockSQLenv
import agent2 as agn
import numpy as np
import utilities as ut
from IPython.display import clear_output

env = mockSQLenv.mockSQLenv(verbose=True)
```

```python
###########################################################3333
##############################################################################################
#restarted with reward 10 for success instead of 100, this should help us avoid getting stuck in suboptimal routes
```

```python

agt = agn.Agent(const.actions)
agt.set_learning_options(exploration=0.02, 
                         learningrate=0.1, 
                         discount=0.9, max_step = 10)
print(agt.Q)
print(agt.total_successes, "/", agt.total_trials)
```

```python
nepisodes = 50
terminated = True
for _ in range(nepisodes):
    clear_output()
    if(terminated):
        env = mockSQLenv.mockSQLenv(verbose=True)
    agt.reset(env)
    
    terminated = agt.run_episode()
    print(ut.getdictshape(agt.Q))
    print(agt.total_successes, "/", agt.total_trials, agt.total_successes/agt.total_trials)
```
